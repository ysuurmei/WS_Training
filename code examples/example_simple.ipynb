# Webscraping Simple Example

Load all required libraries (bs4 for HTML parsing and requests to make HTML requests)

```python
from bs4 import BeautifulSoup
import requests
```

Make an HTML get request to fetch the webpage text using the requests library

```python
url = requests.get('https://www.supermarktaanbiedingen.com/folders/albert_heijn/2017/41')
print(url.text)
```

Parse the page into a BeatifulSoup4 object

```python
source = BeautifulSoup(url.text, 'html.parser')
```

Find all the card containing information on the page, if we print the output, we see it's kind of a mess

```python
cards = source.find_all('div', class_="card")
print(cards)
```

So now let us only extract the card content text and print the output (note the "\n" is only to add an empty new line after each printed statement)

```python
for i in range(len(cards)):
    card_text = cards[i].find('div', class_ = 'card_content').text
    print(card_text, "\n")
```
